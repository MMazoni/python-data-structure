# Fast & Slow Pointers

Essa técnica é comumente utilizada para verificar ciclos em Liked Lists. Inicia-se dois ponteiros no primeiro nó, o `slow` e o `fast`. O slow move de um em um, enquanto o fast move de dois em dois. Se os pointeiros fast e slow se encontrarem na mesma posição, um ciclo foi encontrado.

image.png

### Implementando

```python
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next


def has_cycle(head):
    slow, fast = head, head
    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    
    return False
```

```python
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
print("LinkedList tem um ciclo: " + str(has_cycle(head)))
```
