## Big O notations - Complexidade de Tempo

> T = an + b


### O(1) -> Tempo Constante

Sempre executará no mesmo tempo independente dos dados de entrada

```python
def example():
    print('Hello World') # Esse código terá um tempo de execução constante

example()
```

### O(n) -> Tempo Linear

Quando o tempo do algoritmo é dependente do valor de **n**

```python
def print_items(my_array):
    for item in my_array:  # <-- depende do número de itens no array
        print(item)        # <-- constante

print_items([1,2,3,4,5])
```

### O(log n) -> Tempo Logarítmico

Quando o algoritmo reduz o tamanho dos dados de entrada de cada iteração

**Busca Binária**

image.png

```python
def binary_search(list, item):
    first_index = 0
    last_index = len(list) - 1
    found = False

    while first_index <= last_index and not found:
        midpoint = (first_index + last_index)//2
        if list[midpoint] == item:
            found = True
        else:
            if item < list[midpoint]: # descarta os índices da direita
                last_index = midpoint - 1
            else:
                first_index = midpoint + 1 # descarta os índices da esquerda

    return found

# é necessário ser uma lista ordenada para a busca binária
test_list = [17, 20, 26, 31, 44, 54, 55, 65, 77, 93]
print(binary_search(test_list, 31))
print(binary_search(test_list, 100))
```

### O(n^2) - Tempo Quadrático

Quando o algoritmo tem a necessidade de executar uma operação linear para cada valor dentro dos dados de entrada.



```python
def find_common_elements(list1, list2):
    result = []
    for item1 in list1:
        for item2 in list2:
            if item1 == item2:
                result.append(item1)
    return result

find_common_elements([3,5,2,6], [3,1,2,7])
```

### O(2^n) - Tempo Exponencial

Quando o algoritmo resolve o problema do tamanho **n** recursivamente com 2 sub-problemas e tamanho **n-1**

```python
def fibonacci(num):
    if num <= 1:
        return num
    return fibonacci(num - 2) + fibonacci(num - 1) 

fibonacci(8)
```

### O(n!) - Tempo Fatorial

Quando o tempo do algoritmo cresce de forma fatorial de acordo com os dados de entrada

```python
def factorial(max_number):
    for each in range(max_number):
        print(each)
        factorial(max_number - 1)

factorial(3)
```
