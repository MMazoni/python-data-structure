# Window Sliding

Essa técnica consegue transformar um for loop aninhado em apenas um for, com isso transformando em O(n). Quando tem uma estrutura linear, você criar uma janela de um determinado tamanho. Você irá fazer a "pergunta" se esse conjunto de números encaixa na condição, com isso vai movendo a janela para a direita e fazendo a mesma pergunta. Quando ela chegar no final acaba o loop.


**Quando utilizar?**

- Percorrer uma lista sequencial
- Quer o maior, o menor, o máximo, o mínimo, conter...

### Maximum Sum of 3 consecutive elements of the array

```python
def find_max_sum_subarray(arr, k):
    max_sum = 0
    current_running_sum = 0

    for i in range(len(arr)):
        current_running_sum += arr[i]
        if i >= k - 1:
            max_sum = max(max_sum, current_running_sum)
            current_running_sum -= arr[i - (k - 1)]

    return max_sum


find_max_sum_subarray([4,2,1,7,8,1,2,8,1,0], 3)
```

### Longest Substring without repeating characters



Data uma string, retorne o tamanho da maior substring que não tenha nenhum caracter repetido.

Exemplo 1:

Entrada: abcabcbb Saida: 3

Resposta: O valor encontrado é "abc", que tem o tamanho 3.

Exemplo 2:

Entrada: zzzabcdzzz

Saida: 5

Resposta: O valor encontrado é "zabcd", que tem o tamanho 5.


```python
def solution(s):
    result = 0
    stringSet = set()
    for left in range(len(s)):
        right = left
        while right < len(s) and s[right] not in stringSet:
            stringSet.add(s[right])
            right += 1

        setLength = len(stringSet)
        if setLength > result:
            result = setLength
        stringSet.clear()
    
    return result

solution('abcabcbb')
```
