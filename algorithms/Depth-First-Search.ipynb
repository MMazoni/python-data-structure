# Depth-First Search

É um método para percorrer uma árvore ou grafo. Em DFS, você vai o mais fundo possível em um caminho, antes de voltar e tentar um próximo. 

Passos:

1. começa pela raiz
2. percorre um caminho até encontrar um beco sem saída.
3. faz o mesmo no próximo caminho
4. e assim por diante, até não existir mais caminhos para percorrer

image.png

Vantagens:

- DFS em uma árvore binária, geralmente consume menos memória que BFS.
- DFS pode ser facilmente implementado com recursão.

Desvantagens:

- Não consegue encontrar o caminho mais curto, como BFS.

## Implementando DFS em Árvore Binária

```python
import pdb

class Node:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right


def depth_first_search(node, value_to_find):
    if not node:
        return False
    if node.value == value_to_find:
        return True

    return depth_first_search(node.left, value_to_find) or \
        depth_first_search(node.right, value_to_find)
```

```python
node9 = Node(9)
node8 = Node(8)
node7 = Node(7)
node6 = Node(6)
node5 = Node(5)
node4 = Node(4, node8, node9)
node3 = Node(3, node6, node7)
node2 = Node(2, node4, node5)
node1 = Node(1, node2, node3)
depth_first_search(node1, 9) == True
```

```python
depth_first_search(node1, 0) == False
```
