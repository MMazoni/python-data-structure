# Breadth-First Search

É um algorítmo para atravessar uma árvore ou grafo. Em BFS, você explora primeiro todos nodes em apenas um grau de distância, após isso vai para os nodes com dois graus de distância e assim por diante.

**Passos:**

1. começa pela raiz
2. visita todos os nodes filhos imediatos (em um grau de distância)
3. agora visita os filhos desses nodes
4. assim por diante, até não encontrar mais nodes

image.png

**Vantagens**:

- Um BFS vai encontrar o **caminho mais curto** entre um ponto de início até um determinado nó.

**Desvantagens**:

- Um BFS vai utilizar mais memória que um DFS.

```python
class Node:
    def __init__(self, value=0, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right


def breadth_first_search(root, value_to_find):
    if not root:
        print('There is no root.')
        return False
    
    queue = [root]
    while queue:
        node = queue.pop(0)
        if node.value == value_to_find:
            return True

        if node.left:
            queue.append(node.left)

        if node.right:
            queue.append(node.right)

    return False
    
```
