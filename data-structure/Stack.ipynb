# Stack

É uma estrutura de dados linear que segue o LIFO (Last In First Out), o último que entra na pilha, será o primeiro a sair.
As pilhas tem uma complexidade de tempo 0(1) para inserir, remover, verificar o topo. 

**Quando utilizar?**

- Quando precisamos adicionar ou remover componentes em uma estrutura de dados linear e ela cresce e diminui.
- Quando novos componentes podem ser apenas adicionados e removidos pelo final.
- Histórico de browser, formatação de código em IDEs, pilha de chamada de função, redo e undo de editores.

### Exercício

Dada uma string com apenas os seguintes caracteres '(', ')', '{', '}', '[', ']' determine se uma determinada string é válida.

Uma string é considerada válida se:

- Caracteres de abertura devem ser fechadas pelo mesmo tipo.
- Abertura devem ser fechados com uma chave correspondente.
- Uma string vazia é considerada válida.

```python
def solution(str_to_validate):
    # O(n)
    chars = list(str_to_validate)
    mapped = {
        '[': ']',
        '(': ')',
        '{': '}'
    }
    stack = []
    # O(n)
    while chars:
        value_removed = chars.pop()
        if value_removed in mapped:
            if len(stack) and stack.pop() == mapped[value_removed]:
                continue
        stack.append(value_removed)
    return len(stack) == 0
```

```python
import unittest


class SolutionTest(unittest.TestCase):
    def test_solution_one(self):
        self.assertTrue(solution('[]'))


    def test_solution_two(self):
        self.assertTrue(solution('[()]'))


    def test_solution_three(self):
        self.assertFalse(solution('['))


    def test_solution_four(self):
        self.assertFalse(solution('[('))

    def test_solution_five(self):
        self.assertFalse(solution(')[('))


unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
