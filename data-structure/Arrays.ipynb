```python
array_example = [3,5,7,2,8,5]
```

```python
# Inserir no final

array_example.append(10) # rápido e insere no final do array
array_example
```

```python
# Remover no final

array_example.pop()
array_example
```

```python
# Get the last

array_example[-1]
```

```python
# Get the numbers in the middle

array_example[3:5]
```

```python
# Insert in specific index

array_example.insert(2, 4) # mais lento por ter que ordernar os índices após inserção
array_example
```

```python
# Remover um item específico

array_example.remove(5) # lento como o insert
array_example
```

```python
# Um macete para disabilitar um valor no array

array_example[2] = None
array_example
```

```python
# Outra forma de remover

del array_example[2]
array_example
```

```python
# Sort array

array_example.sort()
array_example
```

```python
sorted([4,0,6,20,8])
```

```python
# Busca binária

import bisect

# O(log n)
bisect.bisect(array_example, 3)
```

```python
# Insere e ordena ao mesmo tempo
import bisect

# O(n log n)
bisect.insort(array_example, 7)
array_example
```

```python
# Iteração em uma lista

squares = [num * num for num in range(10)]
squares
```

```python
# Copiar um array

a = [2,5,2,7,4]
# se eu não fizer essa cópia e passar o 'a' direto, ele vai passar como referẽncia
b = a[:] # poderia ser list(a)
b
```

```python
# Copiar quando é um array de objeto
import copy

a = [{'test': 231, 'aaa': 238}]
b = copy.deepcopy(a)
b
```

```python
# Destruração dos dados

a = [1,2,3,4]
a[2], a[1], a[3], a[0] = a[1], a[2], a[0], a[3]
print(a)
```

```python
my_array = list(range(10))
my_array[::2] # Esse 2 é o step, ele vai para frente de 2 em 2
```

```python
my_array[::-1] # lista ao contrário
```
