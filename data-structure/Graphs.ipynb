# Graphs

Grafo é um estrutura de dados não linear que consiste de nós e bordas. Sua implementação é bem parecido com uma Tree. Essa estrutura resolve diversos problemas reais, como as conexões de amizade de uma rede social.
A vantagem de usar grafos é que ele consegue mostrar muita informação em um formato de fácil entendimento.

**Quando usar?**

- Mecanismos de recomendação
- Rede de comunicação
- Banco de dados
- Grafos sociais, de conhecimento...

### Adjacent List

- Um array ou uma hashtable pode ser usado.
- O tamanho do array é igual o número de vértices.
- Uma entrada `array[i]` representa a lista dos vértices adjacentes para o `iº` vertice.

image.png

```python
class AdjNode:
    def __init__(self, data):
        self.vertex = data
        self.next = None


class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = [None] * self.V

    
    def add_edge(self, source, destination):
        node = AdjNode(destination)
        node.next = self.graph[source]
        self.graph[source] = node

        node = AdjNode(source)
        node.next = self.graph[destination]
        self.graph[destination] = node

    
    def print_graph(self):
        for i in range(self.V):
            print(f"Adjacency list of vertex \n {i}", end="")
            temp = self.graph[i]
            while temp:
                print(f" -> {temp.vertex}", end="")
                temp = temp.next
            print(" \n")
```

```python
V = 7
graph = Graph(V)
graph.add_edge(1, 2)
graph.add_edge(2, 1)
graph.add_edge(2, 3)
graph.add_edge(2, 4)
graph.add_edge(3, 3)
graph.add_edge(4, 1)
graph.add_edge(4, 3)
graph.add_edge(5, 6)

graph.print_graph()
```
