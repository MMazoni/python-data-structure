# Sets

Sets são um conjunto de dados únicos. Pelo menos em Python, a implementação do Set é igual ao de hash, então para saber se um dado está em um set é constante (O(1)). É baseado na teoria de Conjuntos da Matemática.

**Quando utilizar?**

- Uma vantagem de usar Set sobre Lista, é que sua implementação é baseada em hash table, com isso é bem rápido para verificar se existe um item.

image.png

```python
A = {1,2,3,4,5,6}
B = {4,6,7,8,9}

A
```

```python
# Union
print(A | B)
# or 
A.union(B)
```

```python
# Intersection
print(A & B)
# or
A.intersection(B)
```

```python
# Difference
print(A - B)
# or
A.difference(B)
```

```python
# Symmetric difference
print(A ^ B)
# or 
A.symmetric_difference(B)
```

```python
# Transforme dados iteráveis em Set
set(['rock', 'paper', 'scissors'])
```

```python
set('mazoni')
```

```python
# Remove os duplicados e ordena
set([2,5,3,7,8,3,2])
```
