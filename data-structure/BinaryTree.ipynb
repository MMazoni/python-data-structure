# Binary Tree

É uma estutura de dado não linear. Cada nó tem no máximo dois filhos, um na direita outro na esquerda. Ela é chamada de estrutura hierárquica.

A sua busca leva vantagem em cima de arrays por ser 0(log n). Deletar e inserir também são rápidos. Uma desvantagem é caso a árvore esteja desbalanceada (filhos de apenas um lado), aí o ideal é usar Array mesmo.

Na direita temos uma árvore *desbalanceada* e na esquerda uma árvore *balanceada*.

image.png

**Quando utilizar?**

- Encontrar nomes em listas.
- Travessar uma árvore ordenada.
- Encontrar um elemento mais próximo.
- Encontrar todos os elementos menor ou maior que um certo valor

### Implementing a Binary Tree

image.png

```python
import pdb

class BinaryTreeNode:
    left = None
    right = None

    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right        
    

def solution(root):
    if root.left is not None and root.right is not None:
            return root
    root.left, root.right = root.right, root.left
    print(root.value)
    if root.left:
        solution(root.left)
    if root.right:
        solution(root.right)
    return root  
```

```python
node9 = BinaryTreeNode(9)
node8 = BinaryTreeNode(8)
node7 = BinaryTreeNode(7)
node6 = BinaryTreeNode(6)
node5 = BinaryTreeNode(5)
node4 = BinaryTreeNode(4, node8, node9)
node3 = BinaryTreeNode(3, node6, node7)
node2 = BinaryTreeNode(2, node4, node5)
node1 = BinaryTreeNode(1, node2, node3)
solution(node1)
```
